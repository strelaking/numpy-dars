NumPy Yorig'i ArrayNumpy Massivlarini AjratishSplitting-qo'shilishning teskari operatsiyasi.Qo'shilish bir nechta massivlarni biriga birlashtiradi va bo'linish bitta massivni ko'pga ajratadi
.Massivlarni ajratish uchun array_split() dan foydalanamiz, uni ajratmoqchi bo'lgan massivni
va darzlar sonini o'tkazamiz.MisolMassivni 3 qismga bo'ling

```python
import numpy as np
def main():
  a = np.array([1,2,3,4,5,6])
  b = np.array_split(a,4)
  print(b)
main()
```

Eslatma: qaytish qiymati uch tillo o'z ichiga olgan bir qator.Agar massiv talab qilinganidan kamroq elementlarga ega bo'lsa, u
mos ravishda oxiridan moslashadi.Misol

```python
import numpy as np
def main():
  a = np.array([1,2,3,4,5,6])
  b = np.array_split(a,3)
  [print(b[0]),print(b[1]),print(b[2])]
main()
```

Eslatma: biz, shuningdek, usul split bor() mavjud, lekin
elementlar yuqoridagi misol kabi yorig'i uchun manba qator kam bo'lsa, u elementlarni rostlash
bo'lmaydi, array_split() to'g'ri ishlagan, lekin split() muvaffaqiyatsiz edi.Massivlarga Bo'lishArray_split() usulining qaytish qiymati-bu har
bir bo'linishni qator sifatida o'z ichiga olgan qator.Agar siz massivni 3 massivga ajratsangiz, ularga
har qanday massiv elementi singari natijadan kirishingiz mumkin:MisolAjratilgan massivlarga kirish:

2-D massivlarni ajratishda shu sintaksisdan foydalaning.Array_split() usulidan foydalaning, bo'linmoqchi bo'lgan qatorga
va siz qilmoqchi bo'lgan bo'linishlar soniga o'ting.Misol2-D massivni uchta 2 - D massivga ajrating
